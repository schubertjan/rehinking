```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = int(1e5)
position = np.zeros(n) 
current = 10 # we start at island 10
for i in range(n):
    position[i] = current
    #flip a coin
    flip = np.random.choice([-1,1], size = 1)
    proposal = current + flip
    #make sure we stay within the islands
    if  proposal > 10:
        proposal = 1
    if proposal < 1:
        proposal = 10
    # decide
    prob_move = proposal / current
    proposal_choice = np.random.uniform(low = 0, high = 1, size = 1)
    if proposal_choice < prob_move:
        current = proposal
    else:
        current = current
```

```python
fig, ax = plt.subplots(1,2, figsize = (15,8))
ax[0].plot(position[0:100])
pos, cnt = np.unique(position, return_counts=True)
ax[1].bar(pos, cnt)
plt.show()
```
